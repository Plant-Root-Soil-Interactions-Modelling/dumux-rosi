```python
import matplotlib.pyplot as plt
import math
import numpy as np

cL = 5e-7
cD = 0
sC = 50000
sH = 1.02e-6
psi_crit = -5500
psi_x = np.linspace(-15000, -5500, 1000)

alpha = np.exp(-(1-cD)*sC*cL - cD)*np.exp(-sH*(psi_x - psi_crit))

plt.plot(psi_x, alpha)
plt.xlabel("Xylem pressure (cm)")
plt.ylabel("$alpha$")
plt.title("cL = 5e-7")
plt.show()    
```
