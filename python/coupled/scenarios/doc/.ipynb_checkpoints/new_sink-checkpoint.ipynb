# Development of a new sink term

The new sink term is based in combining two approaches

1. We use the steady rate approach to approximate depletion in the rhizosphere, i.e. to obtain a realistic potential at the root soil interface

2. We upscale the root system, by connecting total root surface per layer directly to the root collar, adjusting conductivities

## 1. Steady rate approach

The following diagram explains the 1d axisymmetric geometry, green area is root, yellow area is soil domain. Water movement is dependent on the xylem matric potential $\psi_x$, the potential at the soil root interface $\psi_{s,r}$ and the bulk soil potential $\psi_{b,s}$.

```python
display(SVG(filename='geometry.svg'))
```

The fluxes $q_{root}$ and $q_{soil}$ are given by
\begin{eqnarray}
q_{root} &=& 2 \pi r l k_r (\psi_{s,r} - \psi_x) \\
q_{soil} &=& 2 \pi r l B k_s (\psi_{b,s} - \psi_{s,r}),
\end{eqnarray}


where $r$ is the root radius, $l$ root length, geometry factor $B$ $k_s : = k_s(\psi_{s,r}, \psi_{b,s})$
