```python
import numpy as np
import pickle
from plotnine import *
import pandas as pd 

cs = np.array([100,350,600])*1e-6; #example from Dewar2002 [mol mol-1]
TairC = np.array( [-10,10,30]); #[Â°C]
Qlight =np.array([0,250, 500e-6]);#mean absorbed photon irradiance per leaf segment [mol photons m-2 s-1]  
meanPsi =np.array([-20000,-2000,-200])
csv,TairCv,Qlightv,meanPsiv = np.meshgrid(cs,TairC,Qlight, meanPsi)
Qlightv=Qlightv.flatten()
csv=csv.flatten()
meanPsiv=meanPsiv.flatten()
TairCv = TairCv.flatten()

maxrun = len(Qlightv)
print(maxrun)
print(cs,TairC,Qlight, meanPsi)
```

```python
Sic = np.array([np.full(17,0.) for i in range(maxrun)])
for i in range(maxrun):
    with open('Sic_'+repr(i)+'.pkl','rb') as f:
         Sic[i][:] = pickle.load(f)['ST']
```

```python
Sict = Sic.T
df = pd.DataFrame(Sic, columns =['Chl','oi','fwr','sh','p_lcrit',
              'VcmaxrefChl1', 'VcmaxrefChl2',
             'a1','a3', 'alpha','theta',
             'gamma0', 'gamma1', 'gamma2','g0',
             'kr','kx']) 
```

```python
df
df_mean = df
```

```python
df['cs'] = np.round(csv*1e6)
df['TairC'] = TairCv
df['Qlight'] = Qlightv
df['meanPsi'] =meanPsiv
```

```python

(
    ggplot(df)  # What data to use
    + aes(x='TairC', y= "kr",color='Qlight'  )  # What variable to use
    + geom_point()  # Geometric object to use for drawing
    + facet_grid(['meanPsi','cs'])
)
```

```python

(
    ggplot(df)  # What data to use
    + aes(x='TairC', y= "Chl",color='Qlight'  )  # What variable to use
    + geom_point()  # Geometric object to use for drawing
    + facet_grid(['meanPsi','cs'])
    + xlab("leaf rank")
    + ylab("light interception ratio (-)")
    +theme(legend.text = element_text( size=15),
        legend.title = element_text( size=15),
        legend.position="bottom",#c(0.5,0.1),
        #legend.direction = "horizontal",
        legend.box = "vertical",
        panel.grid.minor = element_line(colour="white"),
        panel.grid.major = element_line(colour="white"),
        panel.background = element_rect(fill="white", color="black"),
        strip.text.y =element_text( size=15) ,
        axis.text = element_text( size=17),
        axis.title = element_text( size=21),
        strip.background = element_rect(fill="white"),
        strip.text.x = element_text(size = 21, hjust = 0,face ="bold"),
        strip.placement = "inside",
        legend.margin=margin()) 
)
```

```python
Sic = np.array([np.full(17,0.) for i in range(maxrun)])
for i in range(maxrun):
    with open('Sic_'+repr(i)+'.pkl','rb') as f:
         Sic[i][:] = pickle.load(f)['ST_conf']
Sict = Sic.T
df_conf = pd.DataFrame(Sic, columns =['Chl','oi','fwr','sh','p_lcrit',
              'VcmaxrefChl1', 'VcmaxrefChl2',
             'a1','a3', 'alpha','theta',
             'gamma0', 'gamma1', 'gamma2','g0',
             'kr','kx']) 
```

```python
df_conf
```

```python
df_conf/df_mean
```

```python
dfmean = pd.DataFrame(data=df_mean.mean(),columns=["mean"])
dfmean["maxst"] = (df_mean + df_conf).max()
dfmean["minst"] = (df_mean - df_conf).min()
dfmean['variables']=dfmean.index
dfmeanVj = dfmean
```

```python

(
    ggplot(dfmeanVj)  # What data to use
    + aes(x='variables', y= "mean" )  # What variable to use
    + geom_bar(stat = "identity")  # Geometric object to use for drawing
    + xlab("parameter")
    + geom_errorbar(aes(ymin="minst", ymax="maxst"), width=.2) 
    + ylab("total Sobol sensitivity")
)
```
