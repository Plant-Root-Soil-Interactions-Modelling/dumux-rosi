```python
import plotnine
import os
import pandas as pd
from plotnine import *
import numpy as np
pathresults = "../results/uqrMaster2102_starchDynamic/"
lateDry = "10to18to25to25dry10871/" 
earlyDry = "10to11to18to25dry10870/" 
baseline = "10to18to25to25wet10869/" 
data_file_delimiter = ","
```

```python
os.getcwd()
```

```python

def getKonz(simStart, weather):
    #directory = pathresults +"6to"+str(simStart+7)+"to28"+weather+"/"
    directory = pathresults +"6to28"+weather+"noCS/"
    cst_file = directory+"C_ST.txt"
    print(cst_file)
    largest_column_count = 0
    with open(cst_file, 'r') as temp_f:
        for line in temp_f:
            last_line = line
            column_count = len(last_line.split(data_file_delimiter)) + 1
            if column_count > largest_column_count:
                largest_column_count = column_count
    print(column_count)
    column_names = [i for i in range(0, largest_column_count)]
    cst = pd.read_csv(cst_file, header=None, delimiter=data_file_delimiter, names=column_names)
    return cst
 
def getVal(directory1,filename):
    file_dir = pathresults+directory1
    largest_column_count = 0
    with open(file_dir+filename, 'r') as temp_f:
        for line in temp_f:
            last_line = line
            column_count = len(last_line.split(data_file_delimiter)) + 1
            if column_count > largest_column_count:
                largest_column_count = column_count
    column_names = [i for i in range(0, largest_column_count)]
    print(filename,largest_column_count)
    val_ = pd.read_csv(file_dir+filename, header=None, delimiter=data_file_delimiter, names=column_names,
                 engine='python' )
    
    time = pd.read_csv(file_dir+"time.txt", header=None, delimiter=data_file_delimiter)
    
    val_ = pd.DataFrame(val_, columns = ['cst'])
    val_['time'] = time.iloc[:,0]
    return val_

def getValAll(valName,fileName):
    valearlyDry = getVal(earlyDry,fileName )
    valearlyDry['scenario'] = "drySpell"
    valearlyDry['growth'] = "a) 11-18d"
    vallateDry = getVal(lateDry,fileName)
    vallateDry['scenario'] = "drySpell"
    vallateDry['growth'] = "b) 18-25d"
    valebaseline = getVal(baseline,fileName)
    valebaseline['scenario'] = "baseline"
    valebaseline['growth'] = "b) 18-25d"

    valAll =pd.concat([valearlyDry,vallateDry,valebaseline])
    valAll['var'] = valName
    return(valAll)   
```

```python
cst = getValAll("cst","C_ST.txt")
ots = getValAll("ots","nodesOrganTypes.txt")
cst['ots'] = ots['ots']
```
