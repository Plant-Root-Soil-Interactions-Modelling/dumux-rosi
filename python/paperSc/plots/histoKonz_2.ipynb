```python
import os
import pandas as pd
import numpy as np
pathresults = "../results/lateDry_1440_5_10to25_20mn_0s_130/printData"
#pathresults = "/DUMUXDune27/DUMUX/dumux-rosi/python/paperSc/results/"
data_file_delimiter = ","
```

```python

iiis = ["none","earlyDry","lateDry"]
for iii in iiis:
    pathresults = "../results/1d1dFHung"+str(iii)+"_1440_5_10to25_20mn_0s_128/printData" #/none_1440_5_10to25_20mn_0s_130/printData"
    #pathresults = "../results/1d1dFHunglateDry_1440_5_10to25_20mn_0s_128/printData" #/none_1440_5_10to25_20mn_0s_130/printData"
    file_names = []
    file_namesv = []
    for file_name in os.listdir(pathresults):
        if file_name.startswith("updateWafter") and file_name.endswith("cont1.csv"):
            file_names.append(file_name)
        if file_name.startswith("updateWafter") and file_name.endswith("vol.csv"):
            file_namesv.append(file_name)
    stds = []
    means = []
    for ffi, ff in enumerate(file_names):
        ffv = ff.replace('cont1', 'vol')
        with open(pathresults+"/"+ffv, 'r') as file:
            lines = file.readlines()

            # Check if there are any lines in the file
            if not lines:
                raise ValueError("The file is empty")

            # Get the last line
            last_line = lines[-1].strip()
            # Split the last line by commas and convert to NumPy array
            array_from_last_lineV = np.array(last_line.split(','), dtype=float)
            
        with open(pathresults+"/"+ff, 'r') as file:
            lines = file.readlines()

            # Check if there are any lines in the file
            if not lines:
                raise ValueError("The file is empty")

            # Get the last line
            last_line = lines[-1].strip()

            # Split the last line by commas and convert to NumPy array
            array_from_last_line = np.array(last_line.split(','), dtype=float)/array_from_last_lineV
            stds.append(np.std(array_from_last_line))
            means.append(np.mean(array_from_last_line))
    print(max(np.array(stds)),max(np.array(stds)/np.array(means))*100)

    import matplotlib.pyplot as plt
    data = np.array(stds)/np.array(means)
    # Plotting histogram
    plt.hist(data, bins=5, edgecolor='black')  # Adjust the number of bins as needed
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title('Histogram of Data')
    plt.grid(True)
    plt.show()
```
