```python
import os
import pandas as pd
import numpy as np
pathresults = "../scripts/results/"
#pathresults = "/DUMUXDune27/DUMUX/dumux-rosi/python/paperSc/results/"
data_file_delimiter = ","

def getVal(mypath, filename, header="infer",names = None):
    fullpath = pathresults +mypath+filename
    if names is None:
        cst = pd.read_csv(fullpath,delimiter=data_file_delimiter, header = header).dropna(how= "all", axis=1)
    else:
        cst = pd.read_csv(fullpath,delimiter=data_file_delimiter, header = header, names = names).dropna(how= "all", axis=1)
    return cst


def get_last_line_as_numpy_array(file_path, filename,dtype = float):
    with open(pathresults+file_path+filename, 'r') as file:
        # Read all lines
        lines = file.readlines()

        # Check if there are any lines in the file
        if not lines:
            raise ValueError("The file is empty")

        # Get the last line
        last_line = lines[-1].strip()

        # Split the last line by commas and convert to NumPy array
        array_from_last_line = np.array(last_line.split(','), dtype=dtype)

        return array_from_last_line
```

```python
def printTheData(p2f, scenario, setId):
    tt = getVal(p2f, "time.txt",names= ['time','Qlight'])
    time = tt.iloc[:,0]
    errs = []
    for nc in range(9):
        errs.append(int(max(get_last_line_as_numpy_array(p2f, "content_diff1d3drel"+str(nc)+".csv"))*1000)/1000)
        
    wrsi =  np.array(list(getVal(p2f, "inner_error.csv")[['errWrsi','errWrsiRealInput']].iloc[-1] ))
    timeStep = getVal(p2f, "inner_loopdata.csv")[['dt_inner', 'n_iter']].iloc[-1] 
    timeStep['dt_inner'] *= 24 *60 #*60
    realTime = getVal(p2f, "totalComputetime.txt",names= ['tot','V1','V2','V3'])['tot'].iloc[-1] /3600#/24
    if scenario == "tillage":
        time_ = (int((max(time))*1e4)/1e4)
    else:
        time_ = (int((max(time))*100)/100)
    print(scenario, setId ,'\t', time_,'\t', errs,
         np.array(timeStep*100,dtype=int)/100, 'realTime',int(realTime*1000)/1000,np.array(wrsi*100,dtype=int)/100)
    try:
        if int(setId) == 61:
            print()
    except:
        pass
        
path2file = r'pdef/{}baseline_1440_10to25_20mn_0s_32/'
plevel = [0,1,2,3]


for i, setId  in enumerate(plevel):
    p2f = path2file.format( setId)
    try:
        printTheData(p2f, "pdef", setId)
    except:
        pass
print()
path2file = r'TraiRhizo/paperSc/40{}_1440_{}_10to25_20mn_0s_32/'
scenarios = ["baseline","earlyDry", "lateDry"] 
setIds = [5,44,61]

result_list_comp = [(scenario, str(setId)) for scenario in scenarios for setId in setIds]


for i, pair  in enumerate(result_list_comp):
    scenario, setId = pair[0], pair[1]
    p2f = path2file.format(scenario, setId)
    try:
        printTheData(p2f, scenario, setId)
    except:
        pass
    

#p2f = "TraiRhizo/3p/40customDry_1440_44_10to10_20mn_0s_4/"
#printTheData(p2f, scenario="trtest", setId="")
```
